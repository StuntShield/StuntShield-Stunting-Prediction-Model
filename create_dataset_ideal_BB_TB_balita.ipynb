```python
def create_data(jenis_kelamin,tinggi_badan,berat_badan,status) :
    return {
            "Jenis Kelamin" : jenis_kelamin,
            "Tinggi Badan (cm)" : round(tinggi_badan, 3),
            "Berat Badan (kg)" : round(berat_badan, 3),
            "Status" : status
          }

def create_csv_Berat_Ideal_024_Balita(tinggi_badan,jenis_kelamin,status,batas_bawah,batas_atas):
  path = os.path.join(jenis_kelamin,'ideal-024')
  path = os.path.join(path,status)
  path = os.path.join(path,str(tinggi_badan))
  os.makedirs(path)
  list_ideal_024_path.append(os.path.join(path,'data-ideal-024-balita.csv'))
  with open(os.path.join(path,'data-ideal-024-balita.csv'), mode='w', newline='') as file:
          fieldnames = ["Jenis Kelamin", "Tinggi Badan (cm)","Berat Badan (kg)","Status"]
          writer = csv.DictWriter(file, fieldnames=fieldnames)
          writer.writeheader()
          jml_range = (batas_atas - batas_bawah)*1000
          for i in range(int(jml_range+2)):
            data = create_data(jenis_kelamin,tinggi_badan,batas_bawah,status)
            writer.writerow(data)
            batas_bawah = batas_bawah + 0.001

def create_csv_Berat_Ideal_2460_Balita(tinggi_badan,jenis_kelamin,status,batas_bawah,batas_atas):
  path = os.path.join(jenis_kelamin,'ideal-2460')
  path = os.path.join(path,status)
  path = os.path.join(path,str(tinggi_badan))
  os.makedirs(path)
  list_ideal_2460_path.append(os.path.join(path,'data-ideal-2460-balita.csv'))
  with open(os.path.join(path,'data-ideal-2460-balita.csv'), mode='w', newline='') as file:
          fieldnames = ["Jenis Kelamin", "Tinggi Badan (cm)","Berat Badan (kg)","Status"]
          writer = csv.DictWriter(file, fieldnames=fieldnames)
          writer.writeheader()
          jml_range = (batas_atas - batas_bawah)*1000
          for i in range(int(jml_range+2)):
            data = create_data(jenis_kelamin,tinggi_badan,batas_bawah,status)
            writer.writerow(data)
            batas_bawah = batas_bawah + 0.001

def compress_folder_to_zip(folder_path, zip_path):
    with zipfile.ZipFile(zip_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
        for root, dirs, files in os.walk(folder_path):
            for file in files:
                zipf.write(os.path.join(root, file), os.path.relpath(os.path.join(root, file), folder_path))

```
